```python
for i in range(10):
    f = open(f"products_{i+1}.csv", 'w')
    f.close()
```

```python
columns = []

for i in products:
    columns.append(i)
print(columns)
```

```python
first_slice = 0
second_slice = 10
file_counter = 1

while file_counter <= 10:
    
    writing_data = products[first_slice:second_slice]
    writing_data.to_csv(f'products_{file_counter}.csv', mode='a', index=True, header=columns)
    first_slice += 10
    second_slice += 10
    file_counter += 1
```
