```python
import pandas as pd
import random as rd
```

```python
products = pd.read_csv("products.csv")
```

```python
products
```

```python
random_numbers = []

for i in range(30):
    random_numbers.append(rd.randint(12, 50))
  
```

```python
random_numbers[5]
brandNames = ['Meditex',
"Healixor",
"Pharmora",
"Tranquilix",
"Vivora",
"Neuravex",
"Symbionix",
"Theraflow",
"Cardivex",
"Immunora"]
print(brandNames)
```

```python
for i in range(len(products)):
    qty = random_numbers[rd.randint(0, 28)]
    products.loc[i, "quantity"] = qty
    products.loc[i, "brand_name"] = brandNames[rd.randint(0, len(brandNames) - 1)]
    products.loc[i, "selling_price"] = rd.randint(30000, 100000)
    products.loc[i, "cost_price"] = rd.randint(10000, 30000)
```

```python
products
```

```python
products["brand_name"]
```

```python
for i in range(10):
    print(i)
    f = open(f"products_{i+1}.csv", 'w')
    f.close()
```

```python
columns = []

for i in products:
    columns.append(i)
print(columns)
```

```python
first_slice = 0
second_slice = 11
file_counter = 1

while file_counter <= 10:
    
    writing_data = products[first_slice:second_slice]
    writing_data.to_csv(f'products_{file_counter}.csv', mode='w', index=False, header=columns)
    first_slice += 10
    second_slice += 10
    file_counter += 1
```
